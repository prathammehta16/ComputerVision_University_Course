```python
#log transformations
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/img4.jpg')

c = 255 / np.log(1 + np.max(image))
log_image = c * (np.log(image + 1))

log_image = np.array(log_image, dtype = np.uint8)


plt.imshow(image)
plt.show()
plt.imshow(log_image)
plt.show()
```

```python

import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/img4.jpg')

gamma_two_point_two = np.array(255*(img/255)**2.2,dtype='uint8')

gamma_point_four = np.array(255*(img/255)**0.4,dtype='uint8')

img3 = cv2.hconcat([gamma_two_point_two,gamma_point_four])
cv2_imshow(img3)
cv2.waitKey(0)
```

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/img4.jpg')
(row, col) = img.shape[0:2]
for i in range(row):
	for j in range(col):
		img[i, j] = sum(img[i, j]) * 0.33
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Sum and Subraction
import cv2
import numpy as np
image = cv2.imread('/content/img4.jpg')
matrix = np.ones(image.shape, dtype = "uint8") * 75
added = cv2.add(image, matrix)
cv2.imwrite('result2.jpg', added)
subtracted = cv2.subtract(image, matrix)
cv2.imwrite('result3.jpg', subtracted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Applying AND, OR operations 
import cv2
import numpy as np
image = cv2.imread('/content/img4.jpg')
matrix = np.ones(image.shape, dtype = "uint8") * 75
added = cv2.bitwise_and(image, matrix)
cv2.imwrite('result4.jpg', added)
subtracted = cv2.bitwise_or(image, matrix)
cv2.imwrite('result5.jpg', subtracted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#applying laplacian filter
import cv2
image = cv2.imread('/content/img4.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
result = cv2.Laplacian(gray, cv2.CV_8U, ksize=3)
cv2.imwrite('result6.jpg', result)
```

```python
#median spatial domain filtering
import cv2
image = cv2.imread('/content/img4.jpg')
result = cv2.medianBlur(image, ksize=5)
cv2.imwrite('result.png', result)
```
