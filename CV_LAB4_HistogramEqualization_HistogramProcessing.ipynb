```python
#Plot histogram of your image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/k1.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Compute the histogram of the image
histogram = cv2.calcHist([gray], [0], None, [256], [0, 256])

# Plot the histogram
plt.plot(histogram)
plt.show()
```

```python
#Perform histogram equalization
import cv2

# Load the image
image = cv2.imread('/content/k1.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Perform histogram equalization
equalized = cv2.equalizeHist(gray)

# Save the result
cv2.imwrite('equalized.png', equalized)
```

```python
# Perform contrast stretching. Analyze the results for different values of (r1,s1) and (r2,s2)
import cv2
import numpy as np

# Load the image
image = cv2.imread('image.png')

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define the scaling factors (r1, s1) and (r2, s2)
r1, s1 = 0.2, 0.5
r2, s2 = 0.8, 1.5

# Perform contrast stretching
stretched = cv2.resize(gray, None, fx=s1, fy=s1, interpolation=cv2.INTER_LINEAR)
stretched[stretched < r1] = r1
stretched = cv2.resize(stretched, None, fx=s2, fy=s2, interpolation=cv2.INTER_LINEAR)
stretched[stretched > r2] = r2

# Save the result
cv2.imwrite('stretched.png', stretched)
```
